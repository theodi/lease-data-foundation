# Data Exploration

Basic data exploration experiments and queries

## 1. Setup & Imports

```python
import os
from dotenv import load_dotenv
from pathlib import Path
from pymongo import MongoClient
```

## 2. Configuration

```python
env_path = Path("../.env")
load_dotenv(env_path)

MONGO_URI = os.getenv("MONGO_URI")
MONGO_DATABASE = os.getenv("MONGO_DATABASE")
MONGO_COLLECTION_EXT = os.getenv("MONGO_COLLECTION_EXT")

print(f"Database:    {MONGO_DATABASE}")
print(f"Collection:  {MONGO_COLLECTION_EXT}")
```

## 3. MongoDB Connection

```python
client = MongoClient(MONGO_URI)
db = client[MONGO_DATABASE]
collection_ext = db[MONGO_COLLECTION_EXT]
```

### Query short leases

```python
filter_query = {"ty": {"$lt": 21}}
filtered_count = collection_ext.count_documents(filter_query)

print(f"Number of leases where tenure years < 21: {filtered_count}")
```

```python
from datetime import datetime, time, timezone

# Get today's date at 00:00:00 UTC
today_midnight = datetime.combine(datetime.now(timezone.utc).date(), time.min, tzinfo=timezone.utc)

query = {
    "ty": {"$lt": 21},
    "exp": {"$gt": today_midnight}  # Greater than today at 00:00:00
}

active_short_leases = collection_ext.count_documents(query)

print(f"Number of not-expired leases where tenure years < 21: {active_short_leases}")
```
